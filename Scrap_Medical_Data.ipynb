```python
for i in String:
    url = 'https://www.mayoclinic.org/diseases-conditions/index?letter='+i
    #request allow you to send HTTP request
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    data = soup.find_all("div", class_="cmp-back-to-top-container__children")[0].find_all("li")
    for entry in data:
        url1 = entry.find("a")
        urls.append(url1)

        disease_name = entry.find("a").text.strip()
        D_name.append(disease_name)
```

```python
for i in range(len(urls)):
    d_url = urls[i]['href']
    print(d_url)

    temp_data = {"Diesase_Name":D_name[i]}
    response = requests.get(d_url)
    soup = BeautifulSoup(response.content, 'html.parser')

    data = soup.find_all("div", class_="content")

    if len(data) <= 0:
        continue
    data = data[0]

    description_element = soup.find("h2", text = "Overview")
    if(description_element):
      description = description_element.find_next("p").text
      temp_data["overview"] = description
    # D_Description.append(description)

    symptoms_element = soup.find("h2", text="Symptoms")

    if(symptoms_element):
      symptoms_content = symptoms_element.find_next("ul").get_text(strip=True)
      temp_data["symptoms_elements"] = symptoms_content
    # L_Symptoms.append(symptoms_content)

    symptoms_element = soup.find("h2", text="Symptoms")
    if(symptoms_element):
      symptoms_paragraph = symptoms_element.find_next("p").text
      temp_data["symptoms_paragraph"] = symptoms_paragraph
    # Symptoms.append(symptoms_paragraph)

    final_data.append(temp_data)
```

```python
import re
def camelcase_to_spaces(text):
  spaced_text = re.sub('([a-z0-9])([A-Z])', r'\1 \2', text)
  return spaced_text

df['symptoms_elements'] = df['symptoms_elements'].apply(lambda x: camelcase_to_spaces(x) if isinstance(x, str) else x)
df['symptoms_paragraph'] = df['symptoms_paragraph'].apply(lambda x: camelcase_to_spaces(x) if isinstance(x, str) else x)
```

```python
def split_symptoms(symptoms_text):
  symptoms_list = symptoms_text.split('.')
  cleaned_symptoms = [symptom.strip() for symptom in symptoms_list if symptom.strip()]
  return cleaned_symptoms

df['symptoms_elements'] = df['symptoms_elements'].apply(lambda x: split_symptoms(x) if isinstance(x, str) else x)
```
